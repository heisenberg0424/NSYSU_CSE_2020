```python
import tensorflow.keras as tf
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X=np.linspace(-7,7,300)
X_test=np.linspace(-7,7,300)
y=X**3+2*(X**2)-3*X-1
y_noise=y+np.random.randint(low=-30,high=30,size=300)
plt.scatter(X,y_noise)
plt.plot(X,y,'r-')
```

```python
#1 layer, 10 units
model=tf.models.Sequential()
model.add(tf.layers.Dense(10,activation='sigmoid',input_dim=1))
#model.add(tf.layers.Dense(10,activation='sigmoid'))
model.add(tf.layers.Dense(1))
adam=tf.optimizers.Adam(lr=0.01)
model.compile(loss='mse',optimizer=adam)
model.fit(X,y,epochs=500,verbose=0)

y_pred=model.predict(X_test)
plt.scatter(X,y_noise)
plt.plot(X,y,'r-',label='Actual')
plt.plot(X_test,y_pred,'y-',label='Predict')
plt.legend()
```

```python
#2 layer, 10 units
model=tf.models.Sequential()
model.add(tf.layers.Dense(10,activation='sigmoid',input_dim=1))
model.add(tf.layers.Dense(5,activation='sigmoid'))
model.add(tf.layers.Dense(1))
adam=tf.optimizers.Adam(lr=0.01)
model.compile(loss='mse',optimizer=adam)
model.fit(X,y,epochs=500,verbose=0)

y_pred=model.predict(X_test)
plt.scatter(X,y_noise)
plt.plot(X,y,'r-',label='Actual')
plt.plot(X_test,y_pred,'y-',label='Predict')
plt.legend()
```

```python
from sklearn import datasets
```

```python
circles_data,circles_data_labels=datasets.make_circles(n_samples=50,factor=0.1,noise=0.1)
for i in range(len(circles_data)):
    if circles_data_labels[i]:
        plt.plot(circles_data[i][0],circles_data[i][1],'ro')
    else:
        plt.plot(circles_data[i][0],circles_data[i][1],'bo')
```

```python
model=tf.models.Sequential()
model.add(tf.layers.Dense(5,input_dim=2,activation='relu'))
model.add(tf.layers.Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer=adam)
model.fit(circles_data,circles_data_labels,epochs=500,verbose=0)
```

```python
xx=np.linspace(-1.5,1.5,40)
yy=np.linspace(-1.5,1.5,40)
gx,gy=np.meshgrid(xx,yy)
z=model.predict(np.c_[gx.ravel(), gy.ravel()])
z=z.reshape(gx.shape)
plt.contourf(gx,gy,1*(z>=0.5),2,cmap=plt.cm.coolwarm,alpha=0.8)
axes=plt.gca()
axes.set_xlim([-1.5,1.5])
axes.set_ylim([-1.5,1.5])
plt.grid('off')
plt.axis('off')

for i in range(len(circles_data)):
    if circles_data_labels[i]:
        plt.plot(circles_data[i][0],circles_data[i][1],'ro')
    else:
        plt.plot(circles_data[i][0],circles_data[i][1],'bo')
```
